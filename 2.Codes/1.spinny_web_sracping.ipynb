```python
# Loading the required libraries for scraping the spinny
import requests
import openpyxl
import pandas as pd
import json
import time
from tqdm import tqdm
```

```python
# Initialising the variables 
page_number = 2
afc_no = 9
bfc_no = 8
lfc_no = 3
# Creating a list of url from spinny website to scrape the cars data
base_url = f"""https://api.spinny.com/v3/api/listing/v3/?city=hyderabad&product_type=cars&category=used&page={page_number}&show_max_on_assured=true&custom_budget_sort=true&assured_fetched_cars={afc_no}&budget_fetched_cars={bfc_no}&luxury_fetched_cars={lfc_no}&ratio_status=available&size=20&total_count=844&exclude_ids_v2=17646040,18002295&prioritize_filter_listing=true&high_intent_required=true&active_banner=true&is_max_certified=0"""
```

```python
# Feature Lists to store the data
car_id = []
car_make = []
car_model = []
car_variant = []
car_mileage = []
car_make_year = []
car_price = []
car_fuel_type = []
car_no_of_owners = []
car_body_type = []
car_transmission = []
car_transmission_sub_type = []
car_city = []
car_registration_year = []
car_insurance_validity_year = []
car_insurance_validity_month = []
```

```python
# There are 844 total used cars in hyderabad each time a page loads 20 cars are being loaded. Hence we need to run the loop for 43 times
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:110.0) Gecko/20100101 Firefox/110.0',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-US,en;q=0.9',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'Cache-Control': 'max-age=0',
    'TE': 'Trailers'
}

for loop_no in tqdm(range(42)):
    
    url = base_url
    
    data = requests.get(url,headers=headers)
    
    if data.status_code == 200:
        json_data = json.loads(data.content)
        car_results = json_data['results']

        for i in range(len(car_results)):
            car_id.append(car_results[i]['id'])
            car_make.append(car_results[i]['make'])
            car_model.append(car_results[i]['model'])
            car_variant.append(car_results[i]['variant'])
            car_mileage.append(car_results[i]['mileage'])
            car_make_year.append(car_results[i]['make_year'])
            car_price.append(car_results[i]['price'])
            car_fuel_type.append(car_results[i]['fuel_type'])
            car_no_of_owners.append(car_results[i]['no_of_owners'])
            car_body_type.append(car_results[i]['body_type'])
            car_transmission.append(car_results[i]['transmission'])
            car_transmission_sub_type.append(car_results[i]['transmission_sub_type'])
            car_city.append(car_results[i]['city'])
            car_registration_year.append(car_results[i]['registration_year'])
            car_insurance_validity_year.append(car_results[i]['applicable_insurance_validity']['year'])
            car_insurance_validity_month.append(car_results[i]['applicable_insurance_validity']['month'])
    else:
        print(f"Unable to fetch content from the url. The status code is {data.status_code}")
    time.sleep(20)

car_data_lists = {
    'id':car_id,
    'make':car_make,
    'model':car_model,
    'variant':car_variant,
    'mileage':car_mileage,
    'make_year':car_make_year,
    'price':car_price,
    'fuel_type':car_fuel_type,
    'no_of_owners':car_no_of_owners,
    'body_type':car_body_type,
    'transmission':car_transmission,
    'transmission_sub_type':car_transmission_sub_type,
    'city':car_city,
    'registration_year':car_registration_year,
    'insurance_year':car_insurance_validity_year,
    'insurance_month':car_insurance_validity_month
}
car_data_hyderabad = pd.DataFrame(car_data_lists)
```

## User Defined Function

```python
# Function for scraping data
def car_data_spinny(city_name : str, cars_count : int):
    
    # Initialising the variables 
    page_number = 2
    afc_no = 9
    bfc_no = 8
    lfc_no = 3
    
    # Feature Lists to store the data
    car_id = []
    car_make = []
    car_model = []
    car_variant = []
    car_mileage = []
    car_make_year = []
    car_price = []
    car_fuel_type = []
    car_no_of_owners = []
    car_body_type = []
    car_transmission = []
    car_transmission_sub_type = []
    car_city = []
    car_registration_year = []
    car_insurance_validity_year = []
    car_insurance_validity_month = []

    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:110.0) Gecko/20100101 Firefox/110.0',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-US,en;q=0.9',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'Cache-Control': 'max-age=0',
    'TE': 'Trailers'
}
    loop_count = round(cars_count/20,0)
    loop_count = int(loop_count)

    # Looping through the pages
    for loop_no in tqdm(range(loop_count)):
        
        url = (
        f"https://api.spinny.com/v3/api/listing/v3/?city={city_name}&product_type=cars"
        f"&category=used&page={page_number}&show_max_on_assured=true&custom_budget_sort=true"
        f"&assured_fetched_cars={afc_no}&budget_fetched_cars={bfc_no}&luxury_fetched_cars={lfc_no}"
        f"&ratio_status=available&size=20&total_count=844"
        f"&exclude_ids_v2=17646040,18002295&prioritize_filter_listing=true"
        f"&high_intent_required=true&active_banner=true&is_max_certified=0")

        #print(url)
        
        data = requests.get(url,headers=headers)
        
        if data.status_code == 200:
            json_data = json.loads(data.content)
            car_results = json_data['results']
    
            for i in range(len(car_results)):
                car_id.append(car_results[i]['id'])
                car_make.append(car_results[i]['make'])
                car_model.append(car_results[i]['model'])
                car_variant.append(car_results[i]['variant'])
                car_mileage.append(car_results[i]['mileage'])
                car_make_year.append(car_results[i]['make_year'])
                car_price.append(car_results[i]['price'])
                car_fuel_type.append(car_results[i]['fuel_type'])
                car_no_of_owners.append(car_results[i]['no_of_owners'])
                car_body_type.append(car_results[i]['body_type'])
                car_transmission.append(car_results[i]['transmission'])
                car_transmission_sub_type.append(car_results[i]['transmission_sub_type'])
                car_city.append(car_results[i]['city'])
                car_registration_year.append(car_results[i]['registration_year'])
                car_insurance_validity_year.append(car_results[i]['applicable_insurance_validity']['year'])
                car_insurance_validity_month.append(car_results[i]['applicable_insurance_validity']['month'])
        
        else:
            print(f"Unable to fetch content from the url. The status code is {data.status_code}")

        # Incrementing the variables in the url
        page_number += 1 
        afc_no += 10
        bfc_no += 9
        lfc_no += 1
        #print(page_number)

    # Storing the entire lists in a dictionary
    car_data_lists = {
        'id':car_id,
        'make':car_make,
        'model':car_model,
        'variant':car_variant,
        'mileange':car_mileage,
        'make_year':car_make_year,
        'price':car_price,
        'fuel_type':car_fuel_type,
        'no_of_owners':car_no_of_owners,
        'body_type':car_body_type,
        'transmission':car_transmission,
        'transmission_sub_type':car_transmission_sub_type,
        'city':car_city,
        'registration_year':car_registration_year,
        'insurance_year':car_insurance_validity_year,
        'insurance_month':car_insurance_validity_month
    }

    # Creating a pandas dataframe
    car_data = pd.DataFrame(car_data_lists)

    # Storing the dataframe inthe form of excel
    car_data.to_csv(f"D:\\DataScience\\1.Personal Projects\\Projects\\1.Machine Learning Projects\\1.Regression\\1.Spinny Car\\1.Scraped Data\\raw_data_{city_name}.csv",index=False)
```

```python
# Looping through the city list
city_list = ['hyderabad','delhi-ncr','bangalore','mumbai','pune','delhi','gurgaon','noida','ahmedabad','chennai','kolkata','lucknow','jaipur','chandigarh',
             'coimbatore','faridabad','ghaziabad','indore','kochi']
cars_count = [836,1635,1017,579,669,1045,377,554,689,547,366,309,101,180,158,154,349,165,180]

for city_name,cars_count in tqdm(zip(city_list,cars_count)):
    car_data_spinny(city_name,cars_count)
```
