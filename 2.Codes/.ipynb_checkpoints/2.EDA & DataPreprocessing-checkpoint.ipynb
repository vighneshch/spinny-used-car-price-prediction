```python
# Loading the libraries
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
# Loading each file into a dataframe and concatenating
folder_path = "D:\\DataScience\\1.Personal Projects\\Projects\\1.Machine Learning Projects\\1.Regression\\1.Spinny Car\\1.Scraped Data"

# Creating an empty dataframe
df = pd.DataFrame(columns=['id','make','model','mileange','make_year','price','fuel_type','no_of_owners','body_type','transmission','transmission_sub_type','city','registration_year','insurance_year','insurance_month'])

# Looping through the files in the directory and loading as a dataframe
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path,file_name)
        df_temp = pd.read_csv(file_path)
        df = pd.concat([df,df_temp],axis=0)
        
# Displaying the first 5 rows
df.head()
```

```python
valid_years = df['insurance_year'] > 0
valid_months = df['insurance_month'].between(1, 12)
df = df[valid_years & valid_months]
```

```python
# Renaming the mileange feature
df.rename(columns={'mileange':'mileage'},inplace=True,errors='coerce')

# Chanding the dtype of each features to relevant dtype
df['mileage'] = df['mileage'].astype('int',errors='raise')
df['no_of_owners'] = df['no_of_owners'].astype('int',errors='raise')
df['make_year'] = df['make_year'].astype('int',errors='raise')
df['registration_year'] = df['registration_year'].astype('int',errors='raise')
df['insurance_year'] = df['insurance_year'].astype('int64',errors='ignore')
df['insurance_month'] = df['insurance_month'].astype('int64',errors='ignore')
df['day'] = 1
df['insurance_validity'] = pd.to_datetime(df[['insurance_year', 'insurance_month', 'day']].rename(columns={'insurance_year': 'year', 'insurance_month': 'month'}),errors='raise')
```
